```python
header = ["SibSp","Parch","Fare","Cabin","Name","Ticket","PassengerId"]
```

```python
datatrain.drop(columns = header ,inplace = True)
```

```python
datatrain["Age"].fillna(datatrain["Age"].mean(),inplace = True)
```

```python
Sex = pd.get_dummies(datatrain['Sex'],drop_first=True)
```

```python
datatrain["Sex"]=Sex
```

```python
datatrain.head()
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split
import statistics
features = ['Pclass', 'Sex']
X = datatrain[features]
y = datatrain['Survived']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state=11)

lr = LogisticRegression()
titanic_model = lr.fit(X_train, y_train)
```

```python
titanic_model
```

```python
y_pred = titanic_model.predict(X_test)
```

```python
y_pred
```

```python
from sklearn import metrics

print(metrics.accuracy_score(y_test, y_pred))
```
